```

def load_movies():
    csv_path = "CsvFiles/movies.csv"
    csv_data= pd.read_csv(csv_path)

    print("I have found the following colums ", csv_data.columns)
    print(csv_data.values)
def load_youtube_links():
    csv_path = "CsvFiles/ml-youtube.csv"
    csv_data= pd.read_csv(csv_path)
    links = csv_data['youtubeId']
    ids = csv_data['movieId']
    download_path = "Data/"
    for i,link in enumerate (links):
        if i > 8550:
            c_path= os.path.join(download_path, '{}.mp4'.format(ids[i]))
            print(c_path)
            if not path.exists(c_path):
                # url of the video 
                url = "https://www.youtube.com/watch?v="+link
                try: 
                    # creating pafy object of the video 
                    video = pafy.new(url) 
                    # getting best stream 
                    best = video.getbest(preftype = "mp4") 
                    print(video.title)
                    ## you can print it to see the quality of the video
                    print("Quality is " ,best)
                    best.download(filepath=c_path)
                
                except: 
                    print("*****No video was found, skipping the data*****")
            else:
              print("Found the data, skipping the data #{}...".format(i))

    print("I have found the following colums ", csv_data.columns)
    
    print(links)

    print("Done")
load_youtube_links()
```
